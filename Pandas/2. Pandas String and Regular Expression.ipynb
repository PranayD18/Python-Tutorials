#### Pandas program to extract email from a specified column of string type of a given DataFrame.

```python
pd.set_option('display.max_columns', 10)
df = pd.DataFrame({
    'name_email': ['Alberto Franco af@gmail.com','Gino Mcneill gm@yahoo.com','Ryan Parkes rp@abc.io', 'Eesha Hinton', 'Gino Mcneill gm@github.com']
    })
df
```

```python
def find_email(text):
    email = re.findall(r'[\w\.-]+@[\w\.-]+',str(text))
    return ",".join(email)
df['email']=df['name_email'].apply(lambda x: find_email(x))
df
```

```python
df = pd.DataFrame({
    'tweets': ['#Obama says goodbye',
               'Retweets for #cash',
               'A political endorsement in #Indonesia', 
               '1 dog = many #retweets',
               'Just a simple #egg #omlete']
    })

def find_hash(text):
    hword=re.findall(r'(?<=#)\w+',text)
    return " ".join(hword)
df['hash_word']=df['tweets'].apply(lambda x: find_hash(x))
df
```

#### Pandas program to extract word mention someone in tweets using @ from the specified column of a given DataFrame.

```python
df = pd.DataFrame({
    'tweets': ['@Obama says goodbye',
               'Retweets for @cash',
               'A political endorsement in @Indonesia',
               '1 dog = many #retweets',
               'Just a simple #egg']
    })
def find_at_word(text):
    word=re.findall(r'(?<=@)\w+',text)
    return " ".join(word)
df['find_at_word']=df['tweets'].apply(lambda x: find_at_word(x))
df
```

#### Pandas program to extract only number from the specified column of a given DataFrame.

```python
df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'address': ['7277 Surrey Ave.','920 N. Bishop Ave.','9910 Golden Star St.', '25 Dunbar St.', '17 West Livingston Court'],
    'company_phone_no': ['Company1-Phone no. 4695168357','Company2-Phone no. 8088729013','Company3-Phone no. 6204658086', 'Company4-Phone no. 5159530096', 'Company5-Phone no. 9037952371']
})
def find_at_word(text):
    word=re.findall(r'[0-9]',text)
    return " ".join(word)
df['find_at_word']=df['address'].apply(lambda x: find_at_word(x))


def find_phone_number(text):
    word=re.findall(r"\b\d{10}\b",text)
    return " ".join(word)
df['find_phone_number']=df['company_phone_no'].apply(lambda x: find_phone_number(x))


df
```

#### Pandas program to extract year between 1800 to 2200 from the specified column of a given DataFrame.

```python
df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'year': ['year 1800','year 1700','year 2300', 'year 1900', 'year 2200']
    })

def find_year(text):
    result=re.findall(r"\b(18[0-9]{2}|19[0-8][0-9]|199[0-9]|2[01][0-9]{2}|2200)\b\b",text)
    return result

df['year_range']=df['year'].apply(lambda x: find_year(x))
df

# findall(r'\b[^\d\W]+\b', text) - Extract only words
```

#### Pandas program to extract only non alphanumeric characters from the specified column of a given DataFrame.

```python
df = pd.DataFrame({
    'company_code': ['c0001#','c00@0^2','$c0003', 'c0003', '&c0004'],
    'year': ['year 1800','year 1700','year 2300', 'year 1900', 'year 2200']
    })
def find_nonalpha(text):
    result = re.findall("[^A-Za-z0-9 ]",text) #^ indicates NOT
    return result

df['nonalpha']=df['company_code'].apply(lambda x: find_nonalpha(x))
df


#re.findall(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*', text) - Find Punctuations
```

#### Pandas program to check whether two given words present in a specified column of a given DataFrame

```python
df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'address': ['9910 Surrey Ave.','92 N. Bishop Ave.','9910 Golden Star Ave.', '102 Dunbar St.', '17 West Livingston Court']
    })

def test_and_cond(text):
    result = re.findall(r'(?=.*Ave.)(?=.*9910).*', text) 
    return " ".join(result)

df['check_two_words']=df['address'].apply(lambda x : test_and_cond(x))

df
```

```python
#Pandas program to extract date (format: mm-dd-yyyy) from a given column of a given DataFrame.

df = pd.DataFrame({
    'company_code': ['Abcd','EFGF', 'zefsalf', 'sdfslew', 'zekfsdf'],
    'date_of_sale': ['12/05/2002','16/02/1999','05/09/1998','12/02/2022','15/09/1997'],
    'sale_amount': [12348.5, 233331.2, 22.5, 2566552.0, 23.0]
})
def find_valid_dates(dt):
    #format: mm-dd-yyyy
    result = re.findall(r'\b(1[0-2]|0[1-9])/(3[01]|[12][0-9]|0[1-9])/([0-9]{4})\b',dt)
    return result
df['valid_dates']=df['date_of_sale'].apply(lambda dt : find_valid_dates(dt))
df
```

```python
def rev(st):
    s = list(st)
    i = 0
    j = len(s)-1

    while i < j:  
        while (i < len(s) and s[i].isdigit() ):
            i += 1
        while (j >=0 and s[j]. isdigit()):
            j -= 1
        if i < j:
            s[j], s[i] = s[i], s[j]
        i += 1
        j -= 1

    return ''.join(s)

print(rev('1234ABCD5678'))
```
